```python
import math
import random
import matplotlib.pyplot as plt

# Function: MonteCarlo estimation of pi
#The monte carlo estimation of pi is dont by generating random points in a square and checking the ratio of points in the circle to the total number of points. 
# The ratio is then multiplied by 4 to get the value of pi.
# pi = 4*(points in circle/total points)
# The amount of points generated is determined by the user.

def MonteCarlo_w_Plot(n, plotflag):
    x=[]
    y=[]
    for i in range(n):
        x.append(random.random())
        y.append(random.random())

    dist = []
    for i in range(n):
        dist.append(math.sqrt(x[i]**2 + y[i]**2))

    in_circle = []
    for i in range(n):
        if dist[i] <= 1:
            in_circle.append(dist[i])
    # Calculate ratio of points in circle to total points
    ratio = len(in_circle)/n
    # Calculate pi
    pi = 4*ratio
    if plotflag:
        return pi, x, y, dist
    else: 
        return pi, 0, 0, 0

```

```python
#plots points and circle for estimation
def plotMonte(x,y, dist):
    # Plot points
    col = []
    for i in range(len(x)):
        if dist[i] <= 1:
            col.append('b')
        else:
            col.append('r')
    plt.figure(figsize=(8,8))
    plt.scatter(x,y,c=col, s=2)
    # Plot circle
    theta = []
    for i in range(1000):
        theta.append(i*2*math.pi/1000)
    plt.title('Monte Carlo Estimation of Pi at n='+str(len(x)))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('equal')
    plt.show()
    return

#provides a clean template to display the data. 
def printTemplate(n, plotflag):
    print('-----------------------------------')
    print('Monte Carlo Estimation of Pi at n =',n)
    pi_est, x, y, dist = MonteCarlo_w_Plot(n, plotflag)
    print('Estimated value of pi: ',pi_est)
    print('Error: ',(abs(math.pi-pi_est)/math.pi)*100,'%')
    if plotflag:
        plotMonte(x,y, dist)

#driver code
def main():
    printTemplate(10**3, False)
    printTemplate(10**4, True)
    printTemplate(10**5, False)
    printTemplate(10**6, False)
```

```python
main()

#notes:
#The error will change randomly between different runs, due to the distribution of points being random, however, the general trend of as n increases, the error decreases, is still true.
```
