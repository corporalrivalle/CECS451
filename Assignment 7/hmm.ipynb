```python
#Make a Hidden Markov Model taking in a for P(x0), b,c for True/False in the transition matrix, and d,f for True/False in the sensor matrix
#and e for the number of steps
#and calculates the probability of the sensor given a stage



def hmm(a,b,c,d,f,elist):
    p=0 #Answer
    for i in range(len(elist)):
        if i==0:
            if elist[i]:
                p = (d*a)/(d*a+f*(1-a))
            else:
                p = (f*(1-a))/(d*a+f*(1-a))
        else:
            if elist[i]:
                p = (d*(b*p+c*(1-p)))/(d*(b*p+c*(1-p))+f*(1-(b*p+c*(1-p))))
            else:
                p = (f*(1-(b*p+c*(1-p))))/(d*(b*p+c*(1-p))+f*(1-(b*p+c*(1-p))))
    not_p = 1-p
    stringform = "<{p:.4f}, {not_p:.4f}>"
    return stringform.format(p=p, not_p=not_p)

# Transition Model:
# Xt-1 | P(Xt) | Variable
# True | 0.7   | b
# False| 0.3   | c

# Sensor Model:
# Xt | P(e|Xt) | Variable
# True | 0.9   | d
# False| 0.2   | f
```

```python
print("Part D:")
def problem1():
    print("P(X_t | e_1:t) = ",hmm(a=0.5, b=0.7, c=0.3, d=0.9, f=0.2, elist=[True, True]))

def problem2():
    print("P(X_t | e_1:t) = ",hmm(a=0.5, b=0.7, c=0.3, d=0.9, f=0.2, elist=[True, True, False]))


problem1()
problem2()
```
